# This program recreates the Barnes-Hut algorithm using a Quad and Particle class. The algorithm groups objects sufficiently far away as a single particle with a representative center of mass to improve the N-body simulation runtime to scale as NlogN.

### Problem 1

```python
class Particle:
    
    def __init__(self, pos=[0,0], vel=[0,0], mass=1):
        '''Particle constructor'''
        self.pos = np.array(pos, dtype="float32")
        self.vel = np.array(vel, dtype="float32")
        self.mass = np.float(mass)
    
    def get_info(self):
        '''Displays the position, velocity, and mass of the particle.'''
        print("Pos={}, Vel={}, Mass={}".format(self.pos, self.vel, self.mass))
    
    def get_new_step(self, other_ps, h, a):
        '''Uses the leapfrog integration scheme to get the next time step.'''
        
        # Generating new particle to update data 
        new_p = Particle(self.pos[:], self.vel[:], self.mass)
        current_acc = np.zeros(2)
        for p in other_ps:
            r = (new_p.pos - p.pos)
            mag = np.sqrt(r.dot(r))
            current_acc -= p.mass * r / (mag * (mag**2 + a**2))
 
        # Updating new position
        new_p.pos += new_p.vel*h + 0.5*current_acc*h**2
        
        # To update the new velocity, we need to move other particles to new time step 
        acc_others = np.zeros((len(other_ps), 2))
        for i in range(len(other_ps)):
            for j in range(len(other_ps)):
                if i != j: 
                    r = (other_ps[i].pos - other_ps[j].pos)
                    mag = np.sqrt(r.dot(r))
                    acc_others[i] -= other_ps[j].mass * r / (mag * (mag**2 + a**2))
            # Adding the contribution of the particle of concern
            r = (other_ps[i].pos - self.pos[:])
            mag = np.sqrt(r.dot(r))
            acc_others[i] -= self.mass * r / (mag * (mag**2 + a**2))
                
        # Generating new other_ps array to update data 
        new_other_ps = [Particle(p.pos[:], p.vel[:], p.mass) for p in other_ps]
        for i in range(len(new_other_ps)):
            new_other_ps[i].pos += new_other_ps[i].vel*h + 0.5*acc_others[i]*h**2

        # Computing acceleration at new time step 
        new_acc = np.zeros(2)
        for p in new_other_ps:
            r = (new_p.pos - p.pos)
            mag = np.sqrt(r.dot(r))
            new_acc -= p.mass * r / (mag* (mag**2 + a**2))
        
        # Updating new velocity
        new_p.vel += 0.5*h*(current_acc + new_acc)
                
        return new_p
```

```python
class Quad: 
    
    def __init__(self, xmin=0, xmax=1, ymin=0, ymax=1):
        '''Quad constructor'''
        # The lower-left point
        self.x0, self.y0 = float(xmin), float(ymin)
        self.x1, self.y1 = float(xmax), float(ymax)
        self.w, self.h = self.x1-self.x0, self.y1-self.y0
        self.diag = np.sqrt(self.w**2 + self.h**2)
        self.children = None
        self.CoM = None
        
    def contains(self, particle):
        '''Checks if particle is contained in quad.'''
        x, y = particle.pos
        out = x > self.x0+self.w or x < self.x0 or \
            y > self.y0+self.h or y < self.y0
        return not out
    
    def add_particle(self, particle):
        '''Adds particle and subdivides quad as appropriate.'''
        # Case 1: particle not contained in quad
        if not self.contains(particle):
            #print("reached case 1")
            return -1
        
        # Case 2: empty quad  
        if self.children == None:
            # For leaves, self.children is a particle object
            # Otherwise, it's a list of the four subquads
            #print("reached case 2")
            self.children = particle 
            return 0
        
        else:
        # Case 3: One existing particle in quad 
            #print("reached case 3")
            
            # Store the existing point
            old_p = self.children
            
            if type(old_p) == type(Particle()):
                ## Making the subquads
                # Left-bottom
                q1 = Quad(self.x0,(self.x0+self.x1)/2,self.y0,(self.y0+self.y1)/2)
                # Right-bottom
                q2 = Quad((self.x0+self.x1)/2,self.x1,self.y0,(self.y0+self.y1)/2)
                # Left-top
                q3 = Quad(self.x0,(self.x0+self.x1)/2,(self.y0+self.y1)/2,self.y1)
                # Right-top
                q4 = Quad((self.x0+self.x1)/2,self.x1,(self.y0+self.y1)/2,self.y1)
                self.children = [q1, q2, q3, q4]
            
                # This specifies the quad number in which the existing point is placed.  
                old_p_q = 0 
                # Recursively placing into appropriate subquad 
                for n,quad in enumerate(self.children):
                    if quad.add_particle(old_p) >= 0:
                        old_p_q = n + 1
                        break
                # Handling memory issues
                if old_p_q == 0:
                    print("Out of memory!")
                    return -1

                # Doing the same for new particle
                for n,quad in enumerate(self.children):
                    if quad.add_particle(particle) >= 0:
                        return n + 1 
            
            # Case 4: More than one existing particle in quad 
            else: 
                #print("reached case 4")
                for n,quad in enumerate(old_p):
                    if quad.add_particle(particle) >= 0:
                        return n + 1 
            
            print("Case not covered")
            return -1
    
    def get_graph(self, xlim=(0,2), ylim=(0,2)):
        '''Generates graph of QuadTree.'''
        def get_graph_helper(root):
            '''Helper recursive function to plot points and quads.'''
            # Case for plotting points
            if type(root.children) == type(Particle()):
                x,y = root.children.pos
                plt.plot(x, y, '.r')
                
            # Case for plotting quads as rectangles
            elif root.children != None:
                for quad in root.children:
                    ax.add_patch(patches.Rectangle((quad.x0, quad.y0), quad.w, quad.h, fill=False))
                    get_graph_helper(quad)

        plt.figure(figsize=(6,6))
        plt.rc('font', size=14)
        ax = plt.axes(xlim=xlim, ylim=ylim)
        get_graph_helper(self)
        
             
    def prune(self):
        '''Prunes all empty quads and update CoM for non-empty ones.'''
        def get_CoM_particle(p_lst):
            '''Helper function to calculate CoM and returns as a particle.'''
            # Filter out non-particles 
            p_lst = list(filter(lambda p: type(p) == type(Particle()), p_lst))
            CoM_mass = np.sum([p.mass for p in p_lst])
            CoM_pos = np.sum([p.mass * p.pos for p in p_lst], axis=0) / CoM_mass
            CoM_vel = np.sum([p.mass * p.vel for p in p_lst], axis=0) / CoM_mass
            return Particle(CoM_pos, CoM_vel, CoM_mass)
        
        # Case 1: quad has a single particle
        if type(self.children) == type(Particle()):
            self.CoM = self.children 
        # Case 2: quad has subquads 
        elif type(self.children) == type([]):
            for quad in self.children:
                quad.prune()
            self.children = list(filter(lambda q: (q.children != None), self.children))
            self.CoM = get_CoM_particle([child.CoM for child in self.children])
            
    def get_other_ps(self, particle, tol):
        '''Gets the effective CoM particles acting on a given particle.'''
        p_lst = [] 

        # Case 1: Quad contains the particle 
        if self.contains(particle):
            # Recurse if subquads exist 
            if type(self.children) == type([]):
                for child in self.children:
                    p_lst += child.get_other_ps(particle, tol)
            # This is just the particle on concern itself 
            elif type(self.children) == type(Particle()):
                return []
            
        # Case 2: Quad does not contain the particle     
        else:
            # Case 2.1: opening angle met 
            d = np.linalg.norm(particle.pos - self.CoM.pos)
            if max(self.w, self.h)/d < tol:
                p_lst.append(self.CoM)
            # Case 2.2: opening angle not met
            else:
                # Recurse if subquads exist
                if type(self.children) == type([]):
                    for child in self.children:
                        p_lst += child.get_other_ps(particle, tol)
                # Otherwise use the other particle directly 
                elif type(self.children) == type(Particle()):
                    return [self.CoM]

        return p_lst
            
```

### Problem 2

```python
def gen_random_pts(N):
    '''Generates random points within a shrunk unit circle with vels of 0.1 in random directions.'''
    pts = []
    np.random.seed(5)
    x = np.random.uniform(-1, 1, size=N)
    y = np.random.uniform(-np.sqrt(1-x**2), np.sqrt(1-x**2), size=N)
    for i in range(N):
        pts.append(Particle([x[i], y[i]], 0.1*(np.sign(np.random.uniform(size=(2))-0.5)), 1))
    return pts

def set_quad_size(pts,pad=0.1):
    '''Dynamic setting of quad size based on size of distribution.'''
    x0, y0 = pts[0].pos
    x1, y1 = pts[0].pos
    for p in pts[1:]:
        x = p.pos[0];
        y = p.pos[1];
        if x > x1:
            x1 = x;
        elif x < x0:
            x0 = x;
        if y > y1:
            y1 = y;
        elif y < y0:
            y0 = y;
    return x0-pad, x1+pad, y0-pad, y1+pad

def stepper(N, theta, h, a, T):
    '''
    Steps through Barnes hut algorithm. 
    Arguments: N = no. of particles, theta = opening angle, h = time step
               a = softening factor, T = total time steps
    Returns: final list of points
    '''
    # Generate initial points 
    pts = gen_random_pts(N)
    
    for t in range(T+1):
        # Generate root quad
        x0, x1, y0, y1 = set_quad_size(pts)
        root = Quad(x0,x1,y0,y1)
        # Add points to quad
        for p in pts:
            root.add_particle(p)
        
        # Show graph at 100 time step increments 
        if t % 20 == 0:
            root.get_graph((x0,x1), (y0,y1))
            plt.title('T = {}'.format(t))
            plt.show()
        
        # We just want to graph the last time step 
        if t==T: 
            break
            
        root.prune()
        
        # Calculate next time step 
        new_pts = [0]*N
        for n,p in enumerate(pts):
            other_ps = root.get_other_ps(p, theta)
            new_pts[n] = p.get_new_step(other_ps, h, a)
        
        pts = new_pts 
            
    return new_pts
```

```python
N = 100
theta = 1;
a = 0.1
h = 0.01
T = 100

start = time.time()
final_ps = stepper(N, theta, h, a, T)
end = time.time()
print("runtime  = {:.2f} minutes".format(((end - start) / 60.0)))
```

### Problem 3

```python
def calc_KE(pts):
    KE = 0
    for p in pts:
        KE += p.mass * np.linalg.norm(p.vel) ** 2
    return KE / 2

def calc_PE(pts):
    PE = 0
    for i in range(len(pts) - 1):
        for j in range(i+1, len(pts)):
            PE -= pts[i].mass * pts[j].mass / np.linalg.norm(pts[i].pos - pts[j].pos)
    return PE

def stepper(N, theta, h, a, T):
    '''
    Steps through Barnes hut algorithm. 
    Arguments: N = no. of particles, theta = opening angle, h = time step
               a = softening factor, T = total time steps
    Returns: final list of points
    '''
    # Generate initial points 
    pts = gen_random_pts(N)
    E_lst = [] 
    for t in range(T+1):
        E_lst.append(calc_KE(pts) + calc_PE(pts))
        # Generate root quad
        x0, x1, y0, y1 = set_quad_size(pts)
        root = Quad(x0,x1,y0,y1)
        # Add points to quad
        for p in pts:
            root.add_particle(p)
        
        # Show graph at 100 time step increments 
        #if t % 20 == 0:
            #root.get_graph((x0,x1), (y0,y1))
            #plt.title('T = {}'.format(t))
            #plt.show()
        
        # We just want to graph the last time step 
        if t==T: 
            break
            
        root.prune()
        
        # Calculate next time step 
        new_pts = [0]*N
        for n,p in enumerate(pts):
            other_ps = root.get_other_ps(p, theta)
            new_pts[n] = p.get_new_step(other_ps, h, a)
        
        pts = new_pts 
            
    return E_lst
```

```python
N = 100
a = 0.1
h = 0.01
T = 100

E_lst = []
for theta in np.arange(0, 3, 0.5):
    E_lst.append(stepper(N, theta, h, a, T))
```
